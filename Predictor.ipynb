```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Additional machine learning libraries
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
data = pd.read_csv('california.csv')
data = data.dropna()  # Drop missing values
X = data.drop(columns=['ocean_proximity', 'median_house_value'])  # Drop non-numeric columns and label
y = data['median_house_value']  # Label

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

california_results = {}
```

```python
# Initialize and fit the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Make predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
lr_r2 = r2_score(y_test, y_pred_lr)
lr_mse = mean_squared_error(y_test, y_pred_lr)
california_results['Linear Regression'] = {'R²': lr_r2, 'MSE': lr_mse}
```

```python


# Initialize and fit SVR with tuned parameters
svr_model = SVR(kernel='linear', C=100, gamma=1)
svr_model.fit(X_train, y_train)

# Make predictions
y_pred_svr = svr_model.predict(X_test)

# Evaluate the model
svr_r2 = r2_score(y_test, y_pred_svr)
svr_mse = mean_squared_error(y_test, y_pred_svr)
california_results['Support Vector Regression'] = {'R²': svr_r2, 'MSE': svr_mse}
```

```python


# Initialize KNN regressor
knn_model = KNeighborsRegressor(n_neighbors=5)

# Fit the model
knn_model.fit(X_train, y_train)

# Make predictions
y_pred_knn = knn_model.predict(X_test)

# Evaluate the model
knn_r2 = r2_score(y_test, y_pred_knn)
knn_mse = mean_squared_error(y_test, y_pred_knn)
california_results['K-Nearest Neighbor Regression'] = {'R²': knn_r2, 'MSE': knn_mse}
```

```python


# Initialize and fit Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
rf_r2 = r2_score(y_test, y_pred_rf)
rf_mse = mean_squared_error(y_test,y_pred_rf)
california_results['Random Forest Regression'] = {'R²': rf_r2, 'MSE': rf_mse}
```

```python


# Initialize and fit Gradient Boosting Regressor
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train, y_train)

# Make predictions
y_pred_gb = gb_model.predict(X_test)

# Evaluate the model
gb_r2 = r2_score(y_test, y_pred_gb)
gb_mse = mean_squared_error(y_test,y_pred_gb)
california_results['Gradient Boost Regression'] = {'R²': gb_r2, 'MSE': gb_mse}
```

```python
ames_data = pd.read_csv('ames.csv')

# Preprocess the data
# Drop non-numeric columns (object types) and fill missing values
ames_data = ames_data.select_dtypes(exclude=['object'])
ames_data.fillna(ames_data.mean(), inplace=True)

# Separate features and labels
X_ames = ames_data.drop(columns=['SalePrice'])  # 'SalePrice' is the label
y_ames = ames_data['SalePrice']

# Scale the features
scaler = StandardScaler()
X_ames_scaled = scaler.fit_transform(X_ames)

# Split into training and testing data
X_train_ames, X_test_ames, y_train_ames, y_test_ames = train_test_split(X_ames_scaled, y_ames, test_size=0.2, random_state=42)
ames_results = {}
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train_ames, y_train_ames)
y_pred_lr = lr_model.predict(X_test_ames)
lr_r2 = r2_score(y_test_ames, y_pred_lr)
lr_mse = mean_squared_error(y_test_ames, y_pred_lr)
ames_results['Linear Regression'] = {'R²': lr_r2, 'MSE': lr_mse}
```

```python
svr_model = SVR(kernel='linear', C=100, gamma=1)
svr_model.fit(X_train_ames, y_train_ames)
y_pred_svr = svr_model.predict(X_test_ames)
svr_r2 = r2_score(y_test_ames, y_pred_svr)
svr_mse = mean_squared_error(y_test_ames, y_pred_svr)
ames_results['Support Vector Regression'] = {'R²': svr_r2, 'MSE': svr_mse}
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_ames, y_train_ames)
y_pred_knn = knn_model.predict(X_test_ames)
knn_r2 = r2_score(y_test_ames, y_pred_knn)
knn_mse = mean_squared_error(y_test_ames, y_pred_knn)
ames_results['K-Nearest Neighbor Regression'] = {'R²': knn_r2, 'MSE': knn_mse}
```

```python
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train_ames, y_train_ames)
y_pred_rf = rf_model.predict(X_test_ames)
rf_r2 = r2_score(y_test_ames, y_pred_rf)
rf_mse = mean_squared_error(y_test_ames, y_pred_rf)
ames_results['Random Forest Regression'] = {'R²': rf_r2, 'MSE': rf_mse}
```

```python
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train_ames, y_train_ames)
y_pred_gb = gb_model.predict(X_test_ames)
gb_r2 = r2_score(y_test_ames, y_pred_gb)
gb_mse = mean_squared_error(y_test_ames, y_pred_gb)
ames_results['Gradient Boost Regression'] = {'R²': gb_r2, 'MSE': gb_mse}
```

```python
headers = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

# Load the Boston dataset using the specified headers and delimiter from the image
df = pd.read_csv('boston.csv', header=None, delimiter=r"\s+", names=headers)

# Preprocess the data following the steps from the image and document:
# The Boston dataset may already be numeric, so no need to drop object columns.

# Fill any missing values with the mean of their respective columns (if any)
df.fillna(df.mean(), inplace=True)

# Separate features and labels
X_boston = df.drop(columns=['MEDV'])  # 'MEDV' is the target label (median value of homes)
y_boston = df['MEDV']

# Scale the features
scaler = StandardScaler()
X_boston_scaled = scaler.fit_transform(X_boston)

# Split into training and testing data
X_train_boston, X_test_boston, y_train_boston, y_test_boston = train_test_split(X_boston_scaled, y_boston, test_size=0.2, random_state=42)

# Initialize a dictionary to store R² and MSE results for all models
boston_results = {}
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train_boston, y_train_boston)
y_pred_lr = lr_model.predict(X_test_boston)
lr_r2 = r2_score(y_test_boston, y_pred_lr)
lr_mse = mean_squared_error(y_test_boston, y_pred_lr)
boston_results['Linear Regression'] = {'R²': lr_r2, 'MSE': lr_mse}
```

```python
svr_model = SVR(kernel='linear', C=100, gamma=1)
svr_model.fit(X_train_boston, y_train_boston)
y_pred_svr = svr_model.predict(X_test_boston)
svr_r2 = r2_score(y_test_boston, y_pred_svr)
svr_mse = mean_squared_error(y_test_boston, y_pred_svr)
boston_results['Support Vector Regression'] = {'R²': svr_r2, 'MSE': svr_mse}
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_boston, y_train_boston)
y_pred_knn = knn_model.predict(X_test_boston)
knn_r2 = r2_score(y_test_boston, y_pred_knn)
knn_mse = mean_squared_error(y_test_boston, y_pred_knn)
boston_results['K-Nearest Neighbor Regression'] = {'R²': knn_r2, 'MSE': knn_mse}
```

```python
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train_boston, y_train_boston)
y_pred_rf = rf_model.predict(X_test_boston)
rf_r2 = r2_score(y_test_boston, y_pred_rf)
rf_mse = mean_squared_error(y_test_boston, y_pred_rf)
boston_results['Random Forest Regression'] = {'R²': rf_r2, 'MSE': rf_mse}
```

```python
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train_boston, y_train_boston)
y_pred_gb = gb_model.predict(X_test_boston)
gb_r2 = r2_score(y_test_boston, y_pred_gb)
gb_mse = mean_squared_error(y_test_boston, y_pred_gb)
boston_results['Gradient Boost Regression'] = {'R²': gb_r2, 'MSE': gb_mse}
```
